```python
import rdflib
import jsonpickle
from pathlib import Path
from spec_parser.model import Model, Namespace, Class, Property, Vocabulary, Individual, Datatype
```

```python
path = Path("spdx/v3.0/model")
```

```python
model = jsonpickle.decode((path / "model.json").read_text())
```

```python
def parse_data(data, extra):
    return {
        "iri": data.iri,
        "summary": data.summary,
        "description": data.description,
        "metadata": data.metadata,
        extra: data.__dict__.get(extra, {})
    }

profiles = {}
for ns in model.namespaces:
    profiles[ns.name] = {
        "classes": {},
        "properties": {},
        "vocabularies": {},
        "individuals": {},
        "datatypes": {}
    }
    for k, v in ns.classes.items():
        profiles[ns.name]["classes"][k] = parse_data(v, "properties")
    for k, v in ns.properties.items():
        profiles[ns.name]["properties"][k] = parse_data(v, "used_in")
    for k, v in ns.vocabularies.items():
        profiles[ns.name]["vocabularies"][k] = parse_data(v, "entries")
    for k, v in ns.individuals.items():
        profiles[ns.name]["individuals"][k] = parse_data(v, "values")
    for k, v in ns.datatypes.items():
        profiles[ns.name]["datatypes"][k] = parse_data(v, "format")

profiles
```
